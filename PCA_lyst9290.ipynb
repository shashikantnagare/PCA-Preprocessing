```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the breast cancer dataset.
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
# Description of the Cancer dataset
print(cancer['DESCR'])
```

```python
# Creating a dataframe for the cancer data
df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
```

```python
df.head()
```

```python
# Checking the covariance matrix
df.cov()
```

```python
# Calculating eigen vectors and values using numpy function
eig_val, eig_vec = np.linalg.eig(df.cov())
```

```python
# printing out eigen values 
print("Eigen Values:", eig_val)
```

```python
# printing out eigen vectors 
print("Eigen Vectors:", eig_vec)
```

```python
# Creating pairs for eigen vectors and absolute eigen values
eig_pair = [(np.abs(eig_val[i]), eig_vec[:,i]) for i in range (len(eig_val))]
```

```python
eig_pair
```

```python
for i in eig_pair:
    print(i[0])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Creating objects for Standardizing variables
scaler = StandardScaler()
```

```python
# Fitting the object on the dataset
scaler.fit(df)
```

```python
# Transforming the data for standardization
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
# Creating PCA object with six Principal Components
pca = PCA(n_components=6)
```

```python
# Fitting and transforming th data using PCA object.
pca.fit_transform(scaled_data)
```

```python
#  the amount of variation Explained from the PCA components
print(pca.explained_variance_ratio_)
```

```python
# Calculating total variation explained
0.44272026 + 0.18971182 + 0.09393163 + 0.06602135 + 0.05495768 + 0.04024522
```

```python
# Fitting and transforming original data into pca components(x_pca)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
print("Shape of original data:", scaled_data.shape)
```

```python
print("Transformed shape:", x_pca.shape)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'], cmap='plasma')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
```

```python
pca.components_
```

```python
X_std = StandardScaler().fit_transform(df)
pca = PCA().fit(X_std)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('No. of PCA components')
plt.ylabel('Variation explained')
plt.show()
```
